```python
import pandas as pd

ref_data = pd.read_csv('https://raw.githubusercontent.com/IAlibay/fragment-opt-abfe-benchmark/refs/heads/main/free_energy_data/abfe.csv')
ref_data = ref_data[ref_data.System == 'MCL-1']
ref_data['ligand'] = ref_data['ligand_ID'].apply(lambda x: f"lig_{x}")

ref_data = ref_data.groupby('ligand').agg({'calc_dG': ['mean', 'sem'], 'exp_dG': 'first'}).reset_index()


# Rename the columns
ref_data.columns = ['ligand', 'simulation_Alibay2022', 'sem_Alibay2022', 'simulation_exp-Alibay2022']
ref_data["sem_exp-Alibay2022"] = len(ref_data)*[0]
ref_data.to_excel("data.xlsx")
ref_data
```
